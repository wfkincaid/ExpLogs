```python
d = find_file('230320.*A174', exp_type='ODNP_NMR_comp/ODNP', expno='ODNP')
d.reorder('t2', first=False); # the semicolon supresses printed output in jupyter
```

```python
image(d)
```

```python
d.ft('t2',shift=True, pad=2048)
d.ft('ph1', unitary=True)
image(d)
```

```python
d.mean('nScans')
image(d)
```
